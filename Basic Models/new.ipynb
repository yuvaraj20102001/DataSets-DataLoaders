```python
class sample_datasets(Dataset):
    def __init__(self):
        csv=np.loadtxt("heart.csv",delimiter=",",dtype=np.float32,skiprows=1)
        self.x=torch.from_numpy(csv[:,:12])
        self.y=torch.from_numpy(csv[:,12])
        self.z=torch.from_numpy(csv[:,13])
        self.len=csv.shape[0]
    def __getitem__(self, index):
        return(self.x[index],self.y[index],self.z[index])
    def __len__(self):
        return(self.len)
dataset=sample_datasets()
len(dataset)
for i in range(len(dataset)):
    print(dataset[i])
    print(i)
```

```python

data_load=DataLoader(dataset=dataset,sampler=SequentialSampler(dataset),batch_size=2,drop_last=True)

for i in data_load:
    print(i)
```
