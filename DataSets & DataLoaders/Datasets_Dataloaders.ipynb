```python
import random


class Random_sampler(Sampler):
    def __init__(self,data_set):
        self.data_set=data_set
    def __iter__(self):
        array=np.random.randint(0,len(dataset),(1,len(dataset)-1))
        return array
    def __len__(self):
        return len(dataset)
```

```python
def collate(dataset):
    data=[]
    for i in range(len(dataset)):
        x,y,z=dataset.__getitem__(i)
        data.append(tuple([x,y,z]))
    return data
```
